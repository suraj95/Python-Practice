```python
import matplotlib.pyplot as plt
```

```python
x_values = [0,1,2,3,4,5]
squares = [0,1,4,9,16,25]
```

```python
# Line graph

plt.plot(x_values,squares)
plt.show()
```

```python
# Scatter plot

x_values = list(range(1000))
squares = [x**2 for x in x_values]

plt.scatter(x_values,squares)
plt.title("Squares of numbers from 0-999", fontsize = 24)
plt.xlabel("Numbers", fontsize = 18)
plt.ylabel("Squares", fontsize = 18)
plt.show()
```

```python
# Using colormap

# The value used to determine the color of each point is passed to the c argument, 
# and the cmap argument specifies which colormap to use.
plt.scatter(x_values,squares, c = squares, cmap = plt.cm.Blues, edgecolor = 'none', s=10)
plt.title("Squares of numbers from 0-999", fontsize = 24)
plt.xlabel("Numbers", fontsize = 18)
plt.ylabel("Squares", fontsize = 18)
plt.show()
```

```python
# Emphasizing points (first and last point)

plt.scatter(x_values,squares, c = squares, cmap = plt.cm.Blues, edgecolor = 'none', s=10)
plt.scatter(x_values[0],squares[0], c = 'green', cmap = plt.cm.Blues, edgecolor = 'none', s=100)
plt.scatter(x_values[-1],squares[-1], c = 'red', cmap = plt.cm.Blues, edgecolor = 'none', s=100)
plt.title("Squares of numbers from 0-999", fontsize = 24)
plt.xlabel("Numbers", fontsize = 18)
plt.ylabel("Squares", fontsize = 18)
plt.show()
```

```python
# Removing axis

plt.scatter(x_values,squares, c = squares, cmap = plt.cm.Blues, edgecolor = 'none', s=10)
plt.scatter(x_values[0],squares[0], c = 'green', cmap = plt.cm.Blues, edgecolor = 'none', s=100)
plt.scatter(x_values[-1],squares[-1], c = 'red', cmap = plt.cm.Blues, edgecolor = 'none', s=100)
plt.title("Squares of numbers from 0-999", fontsize = 24)
plt.xlabel("Numbers", fontsize = 18)
plt.ylabel("Squares", fontsize = 18)
plt.axes().get_xaxis().set_visible(False)
plt.axes().get_yaxis().set_visible(False)
plt.show()
```

```python
# Multiple Plots

x_values = list(range(11)) 
squares = [x**2 for x in x_values] 
cubes = [x**3 for x in x_values]
doubles = [x*2 for x in x_values]

plt.scatter(x_values, squares, c ='blue', edgecolor='none', s=20)
plt.scatter(x_values, cubes, c ='blue', edgecolor='none', s=20)
plt.axis([0,11,0,1100])
plt.show()
```

```python
# Filling in space between datasets

x_values = list(range(11)) 
squares = [x**2 for x in x_values] 

plt.scatter(x_values, squares, c ='blue', edgecolor='none', s=20)
plt.scatter(x_values, cubes, c ='blue', edgecolor='none', s=20)
plt.axis([0,11,0,1100])
plt.fill_between(x_values,cubes,squares,facecolor='blue',alpha=0.25)
plt.show()
```

```python
# Multiple Plots in one figure (sharing X-axis)

fig, axarr = plt.subplots(2,1, sharex=True) #subplots arguments are (num_rows,num_cols)

axarr[0].scatter(x_values,squares) #axarr is a 2D numpy array
axarr[0].set_title('Squares')

axarr[1].scatter(x_values,cubes)
axarr[1].set_title('Cubes')

plt.show()
```

```python
# Multiple Plots in one figure (sharing Y-axis)

fig, axarr = plt.subplots(1,2, sharey=True) #subplots arguments are (num_rows,num_cols)

axarr[0].scatter(x_values,squares) #axarr is a 2D numpy array
axarr[0].set_title('Squares')

axarr[1].scatter(x_values,cubes)
axarr[1].set_title('Cubes')

plt.show()
```

```python
doubles = [2*x for x in x_values] 

fig, axarr = plt.subplots(1,3, sharey=True) #subplots arguments are (num_rows,num_cols)

axarr[0].scatter(x_values,squares) #axarr is a 2D numpy array
axarr[0].set_title('Squares')

axarr[1].scatter(x_values,cubes)
axarr[1].set_title('Cubes')

axarr[2].scatter(x_values,doubles)
axarr[2].set_title('Doubles')

plt.show()
```

```python
triples = [3*x for x in x_values]

fig, axarr = plt.subplots(2,2, sharex=True, sharey=True) #subplots arguments are (num_rows,num_cols)

axarr[0,0].scatter(x_values,squares) #axarr is a 2D numpy array
axarr[0,0].set_title('Squares') 

axarr[0,1].scatter(x_values,cubes)
axarr[0,1].set_title('Cubes')

axarr[1,0].scatter(x_values,doubles)
axarr[1,0].set_title('Doubles')

axarr[1,1].scatter(x_values,triples)
axarr[1,1].set_title('Triples')

plt.show()
```
